3. Write a function that m when provided the first node of aa singly linked list. Will sort the list by its value from small to large . Do not use built in sort method

```python
class new_node():
    def __init__(self,val=0):
        self.val = val
        self.next = None
        
    def set_next(self,node):
        self.next = node
        return self.next
        
    def print_node(self):
        print(self.val)
        if self.next != None:
            self.next.print_node()
        
def rmv(root,rmv):
    #print(root,rmv)
    temp = root
    while temp.next != None:
        if temp.next.val == rmv.val:
            temp.next = rmv.next
            return True
        temp = temp.next
    print(f'{rmv.val} not found')
    return False

def sort_print(root):
    
    temp_node = root
    while temp_node.next != None:
        if temp_node.val > temp_node.next.val:
            # get the lesser root
            if temp_node == root:
                root = temp_node.next
            temp = temp_node.next
            temp_node.next = temp.next
            temp.next = temp_node
            
        temp_node = temp_node.next
    return root
```

```python
root = new_node(26)
to_del = new_node(2656)

root.set_next(new_node(110)).set_next(new_node(170)).set_next(to_del).set_next(new_node(1210)).set_next(new_node(56)).set_next(new_node(10))

root.print_node()

rmv(root,to_del)
print('\n deleting 2656 \n')
root.print_node()

print('\n sorting \n')
root = sort_print(root)
root.print_node()
```
